## Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')

from IPython.display import IFrame
```

## Load some data

```python
measles_df = pd.read_csv('data/measlescasesbycountrybymonth.csv')
measles_df.head()
```

```python
id_cols =  measles_df.columns[0:4]
var_cols = measles_df.columns[4:]
```

```python
measles_long_df = pd.melt(measles_df,
                          id_vars = id_cols,
                          value_vars = var_cols,
                          var_name = 'month',
                          value_name = 'measles_cases')
                         
```

```python
measles_long_df.head()
```

```python
cases_by_region_and_year = measles_long_df.groupby(['Region', 'Year'])['measles_cases'].sum()
cases_by_region_and_year
```

```python
cases_by_region_and_year.unstack(level=0)
```

```python
cases_by_region_and_year.unstack(level=0).plot(kind='bar', stacked=True)
```

## Plotly Express

* https://plotly.com/python/bar-charts/

### Long form data to stacked bar chart

```python
cases_region_year_df=cases_by_region_and_year.reset_index()

cases_region_year_df.head()
```

```python
px.bar(cases_region_year_df,
       x='Year', 
       y='measles_cases', 
       color='Region')
```

### Wide form data to stacked bar chart

```python
cases_region_year_wide_df = cases_by_region_and_year.unstack(level=0).reset_index()
cases_region_year_wide_df
```

```python
cases_region_year_wide_df.columns[1:]
```

```python
px.bar(cases_region_year_wide_df,
       x='Year',
       y=cases_region_year_wide_df.columns[1:])
```
